Thing i forget --> axis = 1 is for column and axis = 0 is for rows
(You have to specify if you want to delete the row or the column)

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
# inplace = True means that the changes are made to the original data set and not a copy of it. 
# If inplace = False, the changes are made to a new copy of the data set and not the original data set.
```

Some changes to the data

```python
data.diagnosis = [1 if value == 'M' else 0 for value in data.diagnosis] # 1 = Malignant, 0 = Benign
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

see here we dont use inplace as we dont want to change the original data

```python
x = data.drop('diagnosis', axis =1)
y = data['diagnosis']
```

```python
x
```

```python
y
```

Divide into target variable and predictors

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2)
```

Train the model

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

Evaluate

```python
predict = lr.predict(x_test)
predict
```

```python
actual = y_test
```

```python
tolerance = 0.1
acceptable_range = tolerance * actual
```

```python
within_range = np.abs(predict - actual) <= acceptable_range
```

```python
num_correct = np.sum(within_range)
total_predictions = len(predict)
```

```python
print(f"Number of correct predictions within tolerance: {num_correct}")
print(f"Total number of predictions: {total_predictions}")
percentage_correct = (num_correct / total_predictions) * 100
print(f"Percentage of correct predictions within tolerance: {percentage_correct:.2f}%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y_test, predict)

plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc_score(y_test, predict))
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
