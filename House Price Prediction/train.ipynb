```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("housing.csv")
```

Saves only the non-null values in the data object

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis = 1)
y = data['median_house_value']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
train_data.hist(figsize = (15,8))
```

```python
train_data.ocean_proximity.value_counts()
```

Ocean Proximity is not a binary value. So we turn every value into a separate feature and give them true or false value

```python
add_data = pd.get_dummies(train_data.ocean_proximity)
add_data = add_data.astype(int)
```

```python
add_data
```

```python
train_data=train_data.join(add_data).drop(['ocean_proximity'],axis = 1)
```

Feature Engineering

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']

train_data['households_ratio'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['median_house_value'],axis = 1),train_data['median_house_value']

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)

add_data = pd.get_dummies(test_data.ocean_proximity)
add_data = add_data.astype(int)
test_data = test_data.join(add_data).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['households_ratio'] = test_data['total_rooms'] / test_data['households']


```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis = 1),test_data['median_house_value']
```

```python
test_data
```

```python
train_data
```

```python
reg.score(x_test, y_test)
```

```python
reg.coef_
```

```python

predictions = reg.predict(x_test)
actual = y_test
```

```python
tolerance = 0.10
acceptable_range = tolerance * actual
```

```python
within_range = np.abs(predictions - actual) <= acceptable_range
```

```python
num_correct = np.sum(within_range)
total_predictions = len(predictions)
```

```python
print(f"Number of correct predictions within tolerance: {num_correct}")
print(f"Total number of predictions: {total_predictions}")
```
